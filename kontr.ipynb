4. Построить график

7. Определить промежутки, на котором f < 0

```python
import math
from sympy import *
from sympy.plotting import plot as plt
init_printing()
```

```python
# 1 Определить корни
x = Symbol('x')
fx = -5 * x ** 2 + 10 * x - 150 
x1, x2 = solve(fx)
print('Первый корень: ', round(x1, 2),
    '\nВторой корень: ', round(x2, 2))
```

```python
# 2 Найти интервалы, на которых функция возрастает
solve(diff(fx) > 0)
```

```python
# 3 Найти интервалы, на которых функция убывает
solve(diff(fx) < 0)
```

```python
# 5 Вычислить вершину
fx = -5 * x**2 + 10 * x - 150
vx = solve(diff(fx))[0]
vy = -5 * vx**2 + 10 * vx - 150
vx, vy
```

```python
# 6 Определить промежутки, на котором f > 0
if solve(fx > 0) ==False:
    print('График ф-и лежит ниже оси х')
else:
    print(solve(fx > 0))
```

```python
# 7 Определить промежутки, на котором f < 0
solve(diff(fx)<0)
print(solve(fx < 0))
```
